```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import chart_studio.plotly as py
import plotly.graph_objs as go 
import plotly
from plotly import __version__
import cufflinks as cf
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
init_notebook_mode(connected=True)
import os.path
import plotly.express as px
import plotly.graph_objs as go 
cf.go_offline()
```

```python
#Here by using pandas 

countryWise=pd.read_csv('datasets_494766_1273307_country_wise_latest.csv')
dayWise=pd.read_csv('datasets_494766_1273307_day_wise.csv')
world=pd.read_csv('datasets_494766_1273307_worldometer_data.csv')
```

```python
#Here I am creating plots using plotly and saving the images 
class graphs():
    #New cases,New Recovered cases and New Deaths based on time
    worldnew=px.line(dayWise[['New cases','New deaths','New recovered']],width=1000.,height=350,title='Graph based on New cases,New Recovered cases and New Deaths based on day count')
    worldnew.update_layout(plot_bgcolor='white',paper_bgcolor= 'white',)
    if os.path.isfile('D:\\GitHub\\CovidDash\\Final Project\\Images\\worldnew.png'):
        os.remove('D:\\GitHub\\CovidDash\\Final Project\\Images\\worldnew.png')
        worldnew.write_image("D:\\GitHub\\CovidDash\\Final Project\\Images\\worldnew.png")
    else:
        worldnew.write_image("D:\\GitHub\\CovidDash\\Final Project\\Images\\worldnew.png")  

    #Whole World
    #This will be a pie chart with active case, recoved case and death cases

    deaths=dayWise['Deaths'].sum()
    recovered=dayWise['Recovered'].sum()
    active=dayWise['Active'].sum()
    qgrps=['Deaths','Active','Recovered']
    qvals=[deaths,active,recovered]
    colors=['#ff6fa9','#ffd0cc','#00b7cc']
    worldPie=px.pie(names=qgrps,values=qvals,width=500.,height=350,title='Pie chart about details till date')
    worldPie.update_layout(

       plot_bgcolor='white',
       paper_bgcolor= 'white',
    )
    worldPie.update_traces(marker=dict(colors=colors, line=dict(color='#000000')))
    if os.path.isfile('D:\\GitHub\\CovidDash\\Final Project\\Images\\worldPie.png'):
        os.remove('D:\\GitHub\\CovidDash\\Final Project\\Images\\worldPie.png')
        worldPie.write_image("D:\\GitHub\\CovidDash\\Final Project\\Images\\worldPie.png")
    else:
        worldPie.write_image("D:\\GitHub\\CovidDash\\Final Project\\Images\\worldPie.png")  




    #This section will give us the image for the whole world.
    data = dict(
            type = 'choropleth',
            colorscale = 'Blues',
            reversescale = True,
            locations = countryWise['Country/Region'],
            locationmode = "country names",
            z=countryWise['Confirmed'],
            text = countryWise['Country/Region'],
            colorbar = {'title' : 'Total Number of cases'},
            ) 
    layout = dict(title = 'Cofirmed cases Country wise',
                    geo = dict(showframe = False),
                  width=790,
                  height=550,
                  )   
    covidMap=go.Figure(data = [data],layout = layout)

    if os.path.isfile('D:\\GitHub\\CovidDash\\Final Project\\Images\\worldTotal.png'):
        os.remove('D:\\GitHub\\CovidDash\\Final Project\\Images\\worldTotal.png')
        covidMap.write_image("D:\\GitHub\\CovidDash\\Final Project\\Images\\worldTotal.png")
    else:
        covidMap.write_image("D:\\GitHub\\CovidDash\\Final Project\\Images\\worldTotal.png")


    #WHO Regions
    #This section will give us count based on WHO region

    wHOTotal = px.bar(world, x="TotalCases", y="WHO Region",orientation='h',width=1000.,height=350,title='Total Count of confirmed cases based on the WHO Regions')

    wHOTotal.update_layout( plot_bgcolor='white',paper_bgcolor= 'white')
    wHOTotal.update_traces(marker_color='green', marker_line_color='rgb(8,48,107)')
    if os.path.isfile('D:\\GitHub\\CovidDash\\Final Project\\Images\\wHOTotal.png'):
        os.remove('D:\\GitHub\\CovidDash\\Final Project\\Images\\wHOTotal.png')
        wHOTotal.write_image("D:\\GitHub\\CovidDash\\Final Project\\Images\\wHOTotal.png")
    else:
        wHOTotal.write_image("D:\\GitHub\\CovidDash\\Final Project\\Images\\wHOTotal.png")


    #This section will give you the world value
    data = dict(
            type = 'choropleth',
            colorscale = 'Greens',
            reversescale = True,
            locations = countryWise['Country/Region'],
            locationmode = "country names",
            z=countryWise['Confirmed'],
            text = countryWise['Country/Region'],
            colorbar = {'title' : 'Total Number of cases'},
            ) 
    layout = dict(title = 'Cofirmed cases Country wise',
                    geo = dict(showframe = False),
                  width=790,
                  height=550,
                  )   
    covidMapWHO=go.Figure(data = [data],layout = layout)

    if os.path.isfile('D:\\GitHub\\CovidDash\\Final Project\\Images\\worldTotalWHO.png'):
        os.remove('D:\\GitHub\\CovidDash\\Final Project\\Images\\worldTotalWHO.png')
        covidMapWHO.write_image("D:\\GitHub\\CovidDash\\Final Project\\Images\\worldTotalWHO.png")
    else:
        covidMapWHO.write_image("D:\\GitHub\\CovidDash\\Final Project\\Images\\worldTotalWHO.png")


    #this section will give you pie chart according to the total count accross the who regions

    afr=world[world['WHO Region']=='Americas']['TotalCases'].sum()
    eu=world[world['WHO Region']=='Europe']['TotalCases'].sum()
    sE=world[world['WHO Region']=='South-EastAsia']['TotalCases'].sum()
    eM=world[world['WHO Region']=='EasternMediterranean']['TotalCases'].sum()
    Us=world[world['WHO Region']=='Americas']['TotalCases'].sum()
    wP=world[world['WHO Region']=='WesternPacific']['TotalCases'].sum()

    qgrps=['Americas','Europe','South-EastAsia','EasternMediterranean','Americas','WesternPacific']
    qvals=[afr,eu,sE,eM,Us,wP]
    colors=['#ae621f','#141a13','#383508','#c6b29b','#e4eaea','#5a7702']
    worldPieWHO=px.pie(names=qgrps,values=qvals,width=500.,height=350,title='Pie chart about details count in WHO Regions')
    worldPieWHO.update_layout(

       plot_bgcolor='white',
       paper_bgcolor= 'white',
    )
    worldPieWHO.update_traces(marker=dict(colors=colors, line=dict(color='#000000')))

    if os.path.isfile('D:\\GitHub\\CovidDash\\Final Project\\Images\\worldPieWHO.png'):
        os.remove('D:\\GitHub\\CovidDash\\Final Project\\Images\\worldPieWHO.png')
        worldPieWHO.write_image("D:\\GitHub\\CovidDash\\Final Project\\Images\\worldPieWHO.png")
    else:
        worldPieWHO.write_image("D:\\GitHub\\CovidDash\\Final Project\\Images\\worldPieWHO.png")

        
#Live World data

data = dict(
            type = 'choropleth',
            colorscale = 'viridis',
            reversescale = True,
            locations = countryWise['Country/Region'],
            locationmode = "country names",
            z=countryWise['Confirmed'],
            text = countryWise['Country/Region'],
            colorbar = {'title' : 'Total Number of cases'},
            ) 
layout = dict(title = 'Cofirmed cases Country wise',
             geo = dict(showframe = False),
             width=790,
             height=550,
                 )   
covidMapLive=go.Figure(data = [data],layout = layout)
if os.path.isfile('D:\\GitHub\\CovidDash\\Final Project\\Images\\covidMapLive.png'):
    os.remove('D:\\GitHub\\CovidDash\\Final Project\\Images\\covidMapLive.png')
    covidMapLive.write_image("D:\\GitHub\\CovidDash\\Final Project\\Images\\covidMapLive.png")
else:
    covidMapLive.write_image("D:\\GitHub\\CovidDash\\Final Project\\Images\\covidMapLive.png")



   
```

```python
from PyQt5 import QtCore, QtGui, QtWidgets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import chart_studio.plotly as py
import plotly.graph_objs as go 
import plotly
from plotly import __version__
import cufflinks as cf
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
init_notebook_mode(connected=True)
import os.path
import plotly.express as px
import plotly.graph_objs as go 
cf.go_offline()
class Ui_MainWindow(object):
    def setupUi(self, MainWindow):
        MainWindow.setObjectName("MainWindow")
        MainWindow.resize(891, 550)
        MainWindow.setStyleSheet('background-color:#ffffff')
        self.centralwidget = QtWidgets.QWidget(MainWindow)
        self.centralwidget.setObjectName("centralwidget")
        self.label = QtWidgets.QLabel(self.centralwidget)
        self.label.setGeometry(QtCore.QRect(20, 10, 201, 41))
        self.label.setObjectName("label")
        self.worldMap = QtWidgets.QLabel(self.centralwidget)
        self.worldMap.setGeometry(QtCore.QRect(330, 10, 531, 391))
        self.worldMap.setObjectName("worldMap")
        self.country = QtWidgets.QLineEdit(self.centralwidget)
        self.country.setGeometry(QtCore.QRect(60, 100, 161, 20))
        self.country.setObjectName("country")
        self.searchButton = QtWidgets.QPushButton(self.centralwidget)
        self.searchButton.setGeometry(QtCore.QRect(240, 100, 75, 23))
        self.searchButton.setObjectName("searchButton")
        self.details = QtWidgets.QLabel(self.centralwidget)
        self.details.setGeometry(QtCore.QRect(10, 150, 321, 111))
        self.details.setObjectName("details")
        self.region = QtWidgets.QComboBox(self.centralwidget)
        self.region.setGeometry(QtCore.QRect(20, 60, 201, 22))
        self.region.setObjectName("region")
        self.region.addItem("")
        self.region.setItemText(0, "")
        self.region.addItem("")
        self.region.addItem("")
        self.region.addItem("")
        self.searchButton_2 = QtWidgets.QPushButton(self.centralwidget)
        self.searchButton_2.setGeometry(QtCore.QRect(240, 60, 75, 23))
        self.searchButton_2.setObjectName("searchButton_2")
        self.label_2 = QtWidgets.QLabel(self.centralwidget)
        self.label_2.setGeometry(QtCore.QRect(10, 100, 31, 21))
        self.label_2.setObjectName("label_2")
        self.Total = QtWidgets.QLabel(self.centralwidget)
        self.Total.setGeometry(QtCore.QRect(0, 350, 541, 211))
        self.Total.setObjectName("Total")
        self.Died_2 = QtWidgets.QLabel(self.centralwidget)
        self.Died_2.setGeometry(QtCore.QRect(530, 310, 351, 241))
        self.Died_2.setObjectName("Died_2")
        MainWindow.setCentralWidget(self.centralwidget)
        self.menubar = QtWidgets.QMenuBar(MainWindow)
        self.menubar.setGeometry(QtCore.QRect(0, 0, 891, 20))
        self.menubar.setObjectName("menubar")
        MainWindow.setMenuBar(self.menubar)
        self.statusbar = QtWidgets.QStatusBar(MainWindow)
        self.statusbar.setObjectName("statusbar")
        MainWindow.setStatusBar(self.statusbar)
        self.searchButton_2.clicked.connect(self.data_view)
        self.searchButton.clicked.connect(self.livedata)
        self.retranslateUi(MainWindow)
        QtCore.QMetaObject.connectSlotsByName(MainWindow)

    def retranslateUi(self, MainWindow):
        _translate = QtCore.QCoreApplication.translate
        MainWindow.setWindowTitle(_translate("MainWindow", "Covid DashBoard"))
        self.label.setText(_translate("MainWindow", "<html><head/><body><p><span style=\" font-size:14pt;\">COVID 19 Dash Board</span></p></body></html>"))
        self.worldMap.setText(_translate("MainWindow", "<html><head/><body><p align=\"center\"><span style=\" font-size:18pt;\">World MAP</span></p></body></html>"))
        self.searchButton.setText(_translate("MainWindow", "Search"))
        self.details.setText(_translate("MainWindow", "<html><head/><body><p><span style=\" font-size:10pt;\">Total Case :- khdabckdhbckad<br/>Active Case :- khasbkcb</span></p><p><span style=\" font-size:10pt;\">Recovered :- khdabckdhbckad<br/>Deceased :- khasbkcb</span></p></body></html>"))
        self.region.setItemText(1, _translate("MainWindow", "World"))
        self.region.setItemText(2, _translate("MainWindow", "WHO Regions"))
        self.region.setItemText(3, _translate("MainWindow", "Country Wise"))
        self.searchButton_2.setText(_translate("MainWindow", "Search"))
        self.label_2.setText(_translate("MainWindow", "Name"))
        self.Total.setText(_translate("MainWindow", "TextLabel"))
        self.Died_2.setText(_translate("MainWindow", "TextLabel"))
        
    def data_view(self):
        userChoice =self.region.currentText() 
        total=self.count()
        recovered=self.recov()
        dead=self.dead()
        
        if userChoice=='World':
            self.worldMap.setPixmap(QtGui.QPixmap("D:\\GitHub\\CovidDash\\Final Project\\Images\\worldTotal.png"))   
            self.worldMap.setScaledContents(True)
            self.Total.setPixmap(QtGui.QPixmap("D:\\GitHub\\CovidDash\\Final Project\\Images\\worldnew.png"))  
            self.Total.setScaledContents(True)
            self.Died_2.setPixmap(QtGui.QPixmap("D:\\GitHub\\CovidDash\\Final Project\\Images\\worldPie.png"))  
            self.Died_2.setScaledContents(True)
            
            self.details.setText("<html><head/><body><p><span style=\" font-size:10pt;\">Total Case :- "+str(total)+"<br/><span style=\" font-size:10pt;\">Recovered :- "+str(recovered)+"<br/>Deceased :- "+str(dead)+"</span></body></html>")
            
        elif userChoice=='WHO Regions':
            self.Total.setPixmap(QtGui.QPixmap("D:\\GitHub\\CovidDash\\Final Project\\Images\\wHOTotal.png"))  
            self.Total.setScaledContents(True)
            self.worldMap.setPixmap(QtGui.QPixmap("D:\\GitHub\\CovidDash\\Final Project\\Images\\worldTotalWHO.png"))   
            self.worldMap.setScaledContents(True)
            self.Died_2.setPixmap(QtGui.QPixmap("D:\\GitHub\\CovidDash\\Final Project\\Images\\worldPieWHO.png"))  
            self.Died_2.setScaledContents(True)
            self.details.setText("<html><head/><body><p><span style=\" font-size:10pt;\">Total Case :- "+str(total)+"<br/><span style=\" font-size:10pt;\">Recovered :- "+str(recovered)+"<br/>Deceased :- "+str(dead)+"</span></body></html>")
            
            
    def livedata(self):
        userChoice =self.region.currentText()
        if userChoice=='Country Wise':
            self.cityName=self.country.text()   
            recovData=self.countryReov(self.cityName)
            totalCaseCountry=self.countryCase(self.cityName)
            deathCountry=self.countryDeath(self.cityName)
            
            self.liveDataGraph(self.cityName)
            self.Total.setPixmap(QtGui.QPixmap("D:\\GitHub\\CovidDash\\Final Project\\Images\\worldLive.png"))  
            self.Total.setScaledContents(True)
            self.Died_2.setPixmap(QtGui.QPixmap("D:\\GitHub\\CovidDash\\Final Project\\Images\\countryPieWHO.png"))  
            self.Died_2.setScaledContents(True)
            self.worldMap.setPixmap(QtGui.QPixmap("D:\\GitHub\\CovidDash\\Final Project\\Images\\covidMapLive.png"))  
            self.worldMap.setScaledContents(True)
            self.details.setText("<html><head/><body><p><span style=\" font-size:10pt;\">Total Case :- "+str(int(totalCaseCountry))+"<br/><span style=\" font-size:10pt;\">Recovered :- "+str(int(recovData))+"<br/>Deceased :- "+str(int(deathCountry))+"</span></body></html>")
            
        
            
            
   
    def liveDataGraph(self,city):
        self.city=city
        liveData=pd.read_csv('https://covid.ourworldindata.org/data/ecdc/total_cases.csv')
        totalCount=liveData['World'][len(liveData['World'])-1]
        worldLive=px.line(liveData[self.city],width=1000.,height=350,title='Graph based on Number of new cases by day')
        worldLive.update_layout(plot_bgcolor='white',paper_bgcolor= 'white')
        if os.path.isfile('D:\\GitHub\\CovidDash\\Final Project\\Images\\worldLive.png'):
            os.remove('D:\\GitHub\\CovidDash\\Final Project\\Images\\worldLive.png')
            worldLive.write_image("D:\\GitHub\\CovidDash\\Final Project\\Images\\worldLive.png")
        else:
            worldLive.write_image("D:\\GitHub\\CovidDash\\Final Project\\Images\\worldLive.png")
            
         #Pie for counties
        fulldata=pd.read_csv('owid-covid-data.csv')
        deathsCount=len(fulldata[fulldata['location']==city]['total_deaths'])
        deaths=fulldata[fulldata['location']==city]['total_deaths'].iloc[deathsCount-1]
        casescount=len(fulldata[fulldata['location']==city]['total_cases'])
        cases=fulldata[fulldata['location']==city]['total_cases'].iloc[casescount-1]
        recovCSV=pd.read_csv('time_series_covid19_recovered_global.csv')
        recov=recovCSV[recovCSV['Country/Region']==city]['6/27/20']
        recov
        qgrps=['Total Cases','Recovered','Death']
        qvals=[cases,float(recov),deaths]
        colors=['#232653','#ac88bf','#909cc4']
        countryPieWHO=px.pie(names=qgrps,values=qvals,width=500.,height=350,title='Pie chart about details count in Specific Country')
        countryPieWHO.update_layout(

           plot_bgcolor='white',
           paper_bgcolor= 'white',
        )
        countryPieWHO.update_traces(marker=dict(colors=colors, line=dict(color='#000000')))
        if os.path.isfile('D:\\GitHub\\CovidDash\\Final Project\\Images\\countryPieWHO.png'):
            os.remove('D:\\GitHub\\CovidDash\\Final Project\\Images\\countryPieWHO.png')
            countryPieWHO.write_image("D:\\GitHub\\CovidDash\\Final Project\\Images\\countryPieWHO.png")
        else:
            countryPieWHO.write_image("D:\\GitHub\\CovidDash\\Final Project\\Images\\countryPieWHO.png")
            
    def count(self):
        liveData=pd.read_csv('https://covid.ourworldindata.org/data/ecdc/total_cases.csv')
        return(liveData['World'][len(liveData['World'])-1])
    def recov(self):    
        recovCSV=pd.read_csv('time_series_covid19_recovered_global.csv')
        return recovCSV['6/27/20'].sum()   
    def dead(self):
        countryWise=pd.read_csv('datasets_494766_1273307_country_wise_latest.csv')
        return countryWise['Deaths'].sum()
    def countryReov(self,city):
        self.city=city
        recovCSV=pd.read_csv('time_series_covid19_recovered_global.csv')
        recov=recovCSV[recovCSV['Country/Region']==self.city]['6/27/20']
        return float(recov)
    def countryCase(self,city):
        self.city=city
        fulldata=pd.read_csv('owid-covid-data.csv')
        casescount=len(fulldata[fulldata['location']==self.city]['total_cases'])
        case=fulldata[fulldata['location']==self.city]['total_cases'].iloc[casescount-1]
        return case
    def countryDeath(self,city):
        self.city=city
        fulldata=pd.read_csv('owid-covid-data.csv')
        deathsCount=len(fulldata[fulldata['location']==self.city]['total_deaths'])
        death=fulldata[fulldata['location']==self.city]['total_deaths'].iloc[deathsCount-1]
        return death
    
                
            
if __name__ == "__main__":
    import sys
    app = QtWidgets.QApplication(sys.argv)
    MainWindow = QtWidgets.QMainWindow()
    ui = Ui_MainWindow()
    ui.setupUi(MainWindow)
   
    MainWindow.show()
    #web.show()
    sys.exit(app.exec_())        
```
